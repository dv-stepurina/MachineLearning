## 5. Function

Write a function that:
1. Takes as arguments: model class of clustering, its parameters, the name of the parameter for optimization, the range of the parameter values to try.
2. Tries different values from the given parameter range and calculates the `silhouette_score` for each value from the range.
3. Finds out the best value for the parameter in the range.
4. Returns two subplots:


- - the first shows how the `silhouette_score` changes depending on the value of the parameter,
- - the second visualizes the data on a `scatter plot` using the clustering model with the best value of the parameter.

```python
def clustering(model_class,  param_name, range_param, X, fixed_params=None):
    
    silhouette_scores = []
    best_score = -1
    best_param = None
    best_labels = None
    for n in range_param:
        if fixed_params != None:
            params = {**fixed_params, param_name: n}
        else:
             params = {param_name: n}
        model = model_class(**params)
        labels = model.fit_predict(X)
        unique_labels = set(labels) - {-1} if -1 in labels else set(labels)
        if len(unique_labels) > 1:
            try:
                score = silhouette_score(X, labels)
            except ValueError:
                score = -1
        else:
            score = -1
        
        silhouette_scores.append(score)
        if score > best_score:
            best_score = score
            best_param = n
            best_labels = labels
    fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(15,16))
    ax1.plot(range_param, silhouette_scores, 'bo-')
    ax1.set_xlabel(param_name)
    ax1.set_ylabel('Silhouette Score')
    ax1.set_title(f'Silhouette Score vs {param_name}\nBest {param_name}: {best_param} (score={best_score:.3f})')
    ax1.grid(True)
    
    if best_labels is not None and len(set(best_labels)) > 1:
        scatter = ax2.scatter(X.iloc[:, 0], X.iloc[:, 1], c=best_labels, cmap='viridis')
        ax2.set_xlabel(X.columns[0])
        ax2.set_ylabel(X.columns[1])
        ax2.set_title(f'Clustering with {param_name}={best_param}')
        plt.colorbar(scatter, ax=ax2, label='Cluster')
    else:
        ax2.text(0.5, 0.5, 'No valid clustering found', 
                ha='center', va='center')
    
    plt.tight_layout()
    return fig, best_param, best_score 
```

```python
fig_db, best_eps, best_score = clustering(
    model_class=AgglomerativeClustering,
    param_name='n_clusters',
    range_param=[3, 7, 8, 9, 15],
    X=df[['num_commits', 'AVG(diff)']]
)
plt.show()
fig_db, best_eps, best_score
```

```python
fig_db, best_eps, best_score = clustering(
    model_class=DBSCAN,
    fixed_params={'eps': 22},
    param_name='min_samples',
    range_param=[1, 2, 3],
    X=df[['num_commits', 'AVG(diff)']]
)
plt.show()
```
